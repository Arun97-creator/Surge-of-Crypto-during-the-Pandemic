## Applying ARIMA on ETH

```r
#ARIMA applicable on Stationary # visual test for Stationary
hist(ETH.ts)
autoplot(ETH.ts, main = "ETH Closing")
```

```r
# As data skewed towards the right # Not stationary

# 2nd method ACF - first calc lags

lags = window(ETH.ts, start=c(2020,1))
gglagplot(lags)
# Look at Lag1 and Lag16 
# Lag1 is on the line and Lag16 are far away from the line - This means the relationship is weak within the variables
# Lag 7-8 relationship is quite strong this where i have to apply  ACF & PACF to reduce the relationship we can't eliminate
# only reduce
```

```r
#ACF of ETH
ggAcf(lags)
```

```r
# ADF TEST Stationarity
#Null hypothesis H0 : Time series is non--stationary
#Alternative Hypotheis HA : Time series is stationary

adf.test(ETH.ts)

# As p is very high we cannot reject the null-hypothesis so It is non-stationary
```

```r
##Non-stationary to stationary
# Differencing
# Remove the unequal variance and remove the trending component
# Now first differencing
ETH_d1 = diff(ETH.ts)
autoplot(ETH_d1,main="First Order Differencing")
```

```r
# Now again To Hypothesis Testing for stationarity
# ADF TEST Stationarity
#Null hypothesis H0 : Time series is non--stationary
#Alternative Hypotheis HA : Time series is stationary
adf.test(ETH_d1,alternative ="stationary", k=0)
# Here p<0.05 so reject the non-stationary H0 Hypothesis and accept stationarity Hypothesis
```

```r
#Second order differencing
ETH_d2 = diff(ETH_d1)
autoplot(ETH_d2,main="Second Order Differencing")
```

```r
# Perform to check stationary and non-stationarity
acf(ETH_d1)
pacf(ETH_d1)
```

```r
# Training and Testing data for ARIMA prediction
tsdatatrain = window(ETH.ts, end=c(2020,10),frequency=365.25) # Train Data set
tsdatatest = window(ETH.ts, end=c(2020,11),frequency=365.25) # Test Data set

tsdatatrain
```

```r
#Auto ARIMA model for ETH
autoarima.fit.train = auto.arima(tsdatatrain,seasonal =TRUE, ic="bic")
autoarima.fit.train
#It says ARIMA (0,1,0) wld be a good fit for ETH
```

```r
# Compare Forecast with Actual Values
autoarima.fit.train.forecast = forecast(autoarima.fit.train, h=10)
autoplot(autoarima.fit.train.forecast)

#Actual vs Forecasted and plotting it
#ETHbind = cbind(tsdatatest,as.data.frame(autoarima.fit.train.forecast)[,4])
#ETHbind


```
