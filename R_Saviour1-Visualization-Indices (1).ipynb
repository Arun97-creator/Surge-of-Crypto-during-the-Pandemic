```r
Tokyo <- getSymbols(Symbols = "^N225",auto.assign = FALSE, from = '2020-01-01', to = '2021-03-08')
HongKong <- getSymbols(Symbols = "^HSI",auto.assign = FALSE, from = '2020-01-01', to = '2021-03-08')
#Shangai <- getSymbols(Symbols = "^399001.SZ",auto.assign = FALSE, from = '2020-01-01', to = '2021-03-08')
Taiwaan <- getSymbols(Symbols = "^TWII",auto.assign = FALSE, from = '2020-01-01', to = '2021-03-08')
#Singapore1 <- na.omit(Singapore)
Korean <- getSymbols(Symbols = "^KS11",auto.assign = FALSE, from = '2020-01-01', to = '2021-03-08')
#NewZealand1 <- na.omit(NewZealand)
```

```r
combined_all_indices1 <- data.frame(date = index(Tokyo), 
                             Tokyo, 
                             row.names = NULL) %>% 
  select(date, close = N225.Close) %>% 
  mutate(ticker = '^N225') %>%
  bind_rows(., 
            data.frame(date = index(HongKong), 
                       HongKong, 
                       row.names = NULL) %>% 
              select(date, close = HSI.Close) %>% 
              mutate(ticker = '^HSI')) %>% 
  bind_rows(., 
            data.frame(date = index(Taiwaan), 
                       Taiwaan, 
                       row.names = NULL) %>% 
              select(date, close = TWII.Close) %>% 
              mutate(ticker = '^TWII')) %>% 
  bind_rows(., 
            data.frame(date = index(Korean), 
                       Korean, 
                       row.names = NULL) %>% 
              select(date, close = KS11.Close) %>% 
              mutate(ticker = '^KS11'))
```

```r
tail(combined_all_indices1)
```

```r
g2 <- ggplot(data = combined_all_indices1, 
            aes(x = date, y = close, color = ticker)) + 
  geom_line()

g2 <- g2 + labs(title = "East-Asia", 
              caption = "Source: Yahoo Finance")

plot(g2)
```

```r
theme_bare1 <- theme(panel.background = element_blank(), 
                    panel.border = element_blank(), 
                    axis.title = element_blank(), 
                    axis.ticks = element_blank(), 
                    panel.grid = element_blank()) 

#choose colors from https://www.schemecolor.com/


stock_colors1 <- c("^N225" = "#e50914", 
                  "^HSI" = "#27076e", 
                  "^TWII" = "#ffb310",
                  "^KS11" = "#86007D")

g2 <- ggplot(data = combined_all_indices1, 
            aes(x = date, y = close, color = ticker)) + 
  geom_line() + scale_color_manual(values = stock_colors1)

g2 <- g2 + theme_bare1 + coord_fixed(ratio = 0.0085) + 
  theme(axis.text = element_text(face = "bold", size = rel(1)), 
        legend.position = "none")

g2 <- g2 + labs(title = "Stock Values at Closing East-Asia", 
              caption = "Source: Yahoo Finance")

g2 <- g2 + annotate("text", 
             x = as.Date('06/12/2020', "%m/%d/%Y"), 
             y = c(20000,30000,10000,3000), 
             label = c("^N225","^HSI","^TWII","^KS11"), 
             color = stock_colors1)

plot(g2)
```
