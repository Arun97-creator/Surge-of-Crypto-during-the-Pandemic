```r
# Making the data more ggplot friendly
Singapore <- data.frame(date = index(Singapore), Singapore, row.names = NULL)

glimpse(Singapore)
```

```r
g1 <- ggplot(data = Singapore, aes(x = date, y = STI.Close)) + 
  geom_line(color = "orange") + 
  scale_x_date(labels = date_format("%Y-%m"), breaks = date_breaks("2 months"))

g1 <- g1 + theme_void() + theme(axis.text = element_text(size = rel(0.8))) 

g1 + coord_fixed(ratio = 5)
g1 + coord_fixed(ratio = 1)
g1 + coord_fixed(ratio = 0.4)
g1 + scale_y_continuous(limits = c(0, 400))


plot(Singapore)
```

```r
combined_all_indices <- data.frame(date = index(Jakarta), 
                             Jakarta, 
                             row.names = NULL) %>% 
  select(date, close = JKSE.Close) %>% 
  mutate(ticker = '^JKSE') %>%
  bind_rows(., 
            data.frame(date = index(Malaysia), 
                       Malaysia, 
                       row.names = NULL) %>% 
              select(date, close = KLSE.Close) %>% 
              mutate(ticker = '^KLSE')) %>% 
  bind_rows(., 
            data.frame(date = index(Australia), 
                       Australia, 
                       row.names = NULL) %>% 
              select(date, close = AXJO.Close) %>% 
              mutate(ticker = '^AXJO')) %>% 
  bind_rows(., 
            data.frame(date = index(Singapore1), 
                       Singapore1, 
                       row.names = NULL) %>% 
              select(date, close = STI.Close) %>% 
              mutate(ticker = '^STI')) %>% 
  bind_rows(., 
            data.frame(date = index(NewZealand1), 
                       NewZealand1, 
                       row.names = NULL) %>% 
              select(date, close = NZ50.Close) %>% 
              mutate(ticker = '^NZ50'))
```
